```python
from skimage import exposure

moon = data.moon()
plt.imshow(moon)
v_min, v_max = np.percentile(moon, (0.2, 99.8))
print(v_min, v_max)
better_contrast = exposure.rescale_intensity(
                                    moon, in_range=(v_min, v_max))
```
